```python
# Binary Thresholding: belirlenen eik deerine gre (thresh value),
# orijinal grntnn pikseli, bu eik deerinden kkse siyah rengine
# dntrlr yani 0 deerini alr, ilgili piksel eik deerinden bykse
# beyaz rengine dntrlr yani 255 deerini alr.

# Binary Inverse (Binary_inv) Thresholding: Binary thresholding'in tam tersidir.
# ilgili piksel, eik deerinden kkse 255 deerini alr (beyaz rengine
# dntrlr), eik deerinden bykse 0 deerini alr (siyah rengine
# dntrlr.

# Trunc Thresholding: Grntnn parlak renklerinin kslmas iin kullanlr.
# yle ki, ilgili piksel deeri eik deerinden bykse eik deerine eitlenir
# eik deerinden kkse ayn braklr.

# TOZERO Thresholding: eer ilgili piksel deeri, eik deerinden bykse, olduu
# gibi korunur yani deimez ayn kalr. Yok eer, piksel deerinden kkse,
# 0 yaplr yani siyah rengine dnr.

# TOZERO Thresholding: TOZERO'nun tam tersidir.
# eer ilgili piksel deeri, eik deerinden bykse, 
# 0 yaplr yani siyah rengine dnr.
# Eer kkse olduu gibi korunur yani deimez ayn kalr.
```

```python
# Python'da 2 adet cv2.resize() fonksiyonu yaznz.
# Bu fonksiyonlar 1 adet parametre alacak. O da grntnn kendisi olacak 
# Grnty bytrken de kltrken de genilik ve yksekliinin
# %25'i kadar olacaktr. Her iki fonksiyon da geriye, ilenmi grntnn
# kendisini dndrecektir.

# cv2.resize() kullanm
# cv2.resize(original_img,(genilik,ykseklik),cv2.INTER_LANCZOS4)
```

```python
def img_zoom_in(img_org):
    # Grntnn satrnn (yksekliinin) %25'ini hesapladk
    new_height = img_org.shape[0] + int(img_org.shape[0]*0.25)
    new_weight = img_org.shape[1] + int(img_org.shape[1]*0.25)

    new_img = cv2.resize(img_org,(new_weight, new_height),cv2.INTER_LANCZOS4)
    return new_img

def img_zoom_out(img_org):
    # Grntnn satrnn (yksekliinin) %25'ini hesapladk
    new_height = img_org.shape[0] - int(img_org.shape[0]*0.25)
    new_weight = img_org.shape[1] - int(img_org.shape[1]*0.25)

    new_img = cv2.resize(img_org,(new_weight, new_height),cv2.INTER_LANCZOS4)
    return new_img
```

```python
my_rgb_img = cv2.cvtColor(my_img, cv2.COLOR_BGR2RGB)
```

```python
new_big_img = img_zoom_in(my_rgb_img)
new_small_img = img_zoom_out(my_rgb_img)

print(f"Orijinal Grnt = {my_rgb_img.shape[:2]}")
print(f"Bytlm Grnt = {new_big_img.shape[:2]}")
print(f"Kltlm Grnt = {new_small_img.shape[:2]}")
```

```python
fig = plt.figure(figsize=(10,7))

plt.subplot(1,3,1)
plt.imshow(my_rgb_img)
plt.axis("off")
plt.title("Original Gray Image")

plt.subplot(1,3,2)
plt.imshow(new_big_img)
plt.axis("off")
plt.title("BIGGER Image")

plt.subplot(1,3,3)
plt.imshow(new_small_img)
plt.axis("off")
plt.title("SMALLER Image")

plt.tight_layout()
plt.savefig("./Image_Resizing_with_automatic_calculation.png",dpi=300)
```

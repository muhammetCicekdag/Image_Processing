```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# İstediğimiz boyutlarda rastgele değerlerden oluşan bir matris tanımlayalım

my_arr_1 = np.random.randint(0,101,size=(5,5),dtype=np.uint8)
```

```python
print(my_arr_1)
print(my_arr_1.shape)
```

```python
height, width, channel = 150, 210, 3

my_black_img = np.zeros((height,width,channel),dtype=np.uint8)
my_white_img = np.ones((height,width,channel),dtype=np.uint8) * 255

# iki farklı yoldan gri rengini oluşturm
my_gray_img_1 = np.random.randint(127,128,size=(height,width,channel),dtype=np.uint8)
my_gray_img_2 = np.ones((height,width,channel),dtype=np.uint8) * 127
```

```python
fig = plt.figure(figsize=(10,7),facecolor="purple")

plt.subplot(2,2,1)
plt.imshow(my_black_img)
plt.axis("off")
plt.title("Black Image")

plt.subplot(2,2,2)
plt.imshow(my_white_img)
plt.axis("off")
plt.title("White Image")

plt.subplot(2,2,3)
plt.imshow(my_gray_img_1)
plt.axis("off")
plt.title("Gray-1 Image")

plt.subplot(2,2,4)
plt.imshow(my_gray_img_2)
plt.axis("off")
plt.title("Gray-2 Image")
```

```python
# 3 kanallı (RGB) bir görüntüde kırmızı rengini [255,0,0]
# bu, şu anlama gelir: Red=255, Green = 0, Blue = 0
# yeşil renginin kodları = [0,255,0]
# mavi renginin kodları = [0,0,255]

# Red, Green, Blue renklerini oluşturup tek bir plot'ta gösterin

my_red_img = np.zeros((height,width,channel),dtype=np.uint8)
my_green_img = np.zeros((height,width,channel),dtype=np.uint8)
my_blue_img = np.zeros((height,width,channel),dtype=np.uint8)
```

```python
my_red_img[:,:,0] = 255
my_green_img[:,:,1] = 255
my_blue_img[:,:,2] = 255
```

```python
fig = plt.figure(figsize=(10,7))

plt.subplot(1,3,1)
plt.imshow(my_red_img)
plt.axis("off")
plt.title("Red Image")

plt.subplot(1,3,2)
plt.imshow(my_green_img)
plt.axis("off")
plt.title("Green Image")

plt.subplot(1,3,3)
plt.imshow(my_blue_img)
plt.axis("off")
plt.title("Blue Image")
```

```python
my_new_black_img = my_red_img * my_green_img * my_blue_img
my_new_white_img = my_red_img + my_green_img + my_blue_img
```

```python
# red =    [255,0,0]
# green =  [0,255,0]
# blue =   [0,0,255]

fig = plt.figure(figsize=(10,7), facecolor="purple")

plt.subplot(1,2,1)
plt.imshow(my_new_black_img)
plt.axis("off")
plt.title("Black Image")

plt.subplot(1,2,2)
plt.imshow(my_new_white_img)
plt.axis("off")
plt.title("White Image")
```

```python
# tanımladığımız my_red, my_green ve my_blue değişkenlerini (matrislerini)
# kullanarak mor ve sarı renklerini elde ediniz

my_yellow_img = my_red_img + my_green_img
my_purple_img = my_red_img + my_blue_img
```

```python
fig = plt.figure(figsize=(10,7))

plt.subplot(1,2,1)
plt.imshow(my_yellow_img)
plt.axis("off")
plt.title("Yellow Image")

plt.subplot(1,2,2)
plt.imshow(my_purple_img)
plt.axis("off")
plt.title("Purple Image")
```

```python
# Fenerbahçe renklerini çiziniz

flag_fenerbahce = np.zeros((height,width,channel),dtype=np.uint8)
```

```python
# matrisin sütununu ikiye böldük ve ilk yarım kısmını sarı renge boyadık
flag_fenerbahce[:,:flag_fenerbahce.shape[1]//2,:] = [255,255,0]

# matrisin ikinci sütununu maviye boyadık
flag_fenerbahce[:,flag_fenerbahce.shape[1]//2:,:] = [0,0,255]

# NOT: eğer bir görseli eşit şekilde parçalayıp (bölmek) farklı renklere
# boyamak istiyorsanız satır veya sütun bazında indeksleme yapmalısınız
# görseli yatay olarak parçalamak ve boyamak istiyorsanız satır indekslemesi
# dikey olarak parçalamak ve boyamak istiyorsanız sütun indekslemesi yapmalısınız
```

```python
plt.imshow(flag_fenerbahce)
plt.axis("off")
plt.title("ŞAMPİYON FENERBAHÇE")
```

```python
# COLOR DETECTION (Renk Tespit Etme): bu işlem için görüntünün tüm
# satır ve sütununda ve/veya istediğimiz  ya da tüm
# kanallarında gezmek durumundayız

# Örneğin, fenerbahçe renklerini gs renklerine dönüştürmek için
# önce fenerbahçe amblemindeki mavi rengini tespit etmemiz lazım
# sonra bunun koordinat sistemini çıkarmamız lazım
# sonra bu koordinat sistemini tekrar bir bütün indeks olarak kullanıp
# bunun değerini kırmızı rengi olarak atamamız lazım

# İlk olarak fenerbahçe amblemindeki mavi rengini tespit ediyoruz
# mavi renginin kodları = [0,0,255]
# yani görüntünün kanalındaki 0. indeksinin değeri = 0
# görüntünün kanalındaki 1. indeksinin değeri = 0
# görüntünün kanalındaki 2. indeksinin değeri = 255

koordinat_mavi = (flag_fenerbahce[:,:,0]==0) & (flag_fenerbahce[:,:,1]==0) & (flag_fenerbahce[:,:,2]==255)
```

```python
# Mavi renginin koordinatlarını tekrar fenerbahçe amblemine yerleştiriyoruz
# Bu defa karşımıza artık koordinat değil, o koordinattaki değerler geliyor
# bu aşamadan sonra da mavi rengi yerine kırmızı renginin kodlarını yazıyoruz
# yani [255,0,0]

flag_fenerbahce[koordinat_mavi] = [255,0,0]
```

```python
plt.imshow(flag_fenerbahce)
plt.axis("off")
plt.title("EN ŞANSLI TAKIM")
```

```python
# g*s amblemindeki sarı rengini beyaza,
# kırmızı rengini de siyaha dönüştürün

# bu defa 2 rengi tespit edip, ikisini de değiştireceksiniz

# ilk olarak sarı rengini tespit ediyoruz

koordinat_sari = (flag_fenerbahce[:,:,0]==255) & (flag_fenerbahce[:,:,1]==255) & (flag_fenerbahce[:,:,2]==0)

koordinat_kirmizi = (flag_fenerbahce[:,:,0]==255) & (flag_fenerbahce[:,:,1]==0) & (flag_fenerbahce[:,:,2]==0)
```

```python
# sarı renginin koordinatlarındaki değerleri beyaz renginin değerlerine dönüştürüyoruz
flag_fenerbahce[koordinat_sari] = [255,255,255]

# kırmızı renginin koordinatlarındaki değerleri siyah renginin değerlerine dönüştürüyoruz
flag_fenerbahce[koordinat_kirmizi] = [0,0,0]
```

```python
fig = plt.figure(figsize=(4,3),facecolor="purple")
plt.imshow(flag_fenerbahce)
plt.axis("off")
plt.title("EN GARİBAN TAKIM")
```
